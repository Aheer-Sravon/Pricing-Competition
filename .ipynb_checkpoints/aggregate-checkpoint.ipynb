```python
import numpy as np
import pandas as pd
import statistics
```

# Q vs Q - Logit

```python
model = 'logit' # row number = 0

avg_prices = {
    "shock_none": [],
    "shock_a": [],
    "shock_b": [],
    "shock_c": []
}

deltas = {
    "shock_none": [],
    "shock_a": [],
    "shock_b": [],
    "shock_c": []
}

rpdis = {
    "shock_none": [],
    "shock_a": [],
    "shock_b": [],
    "shock_c": []
}

for i in range(1, 8):
    # load the dataframes
    df_shock_none = pd.read_csv(f"./q_vs_q/results/run_{i}/shock_none/q_vs_q.csv")
    df_shock_a = pd.read_csv(f"./q_vs_q/results/run_{i}/shock_a/q_vs_q_schemeA.csv")
    df_shock_b = pd.read_csv(f"./q_vs_q/results/run_{i}/shock_b/q_vs_q_schemeB.csv")
    df_shock_c = pd.read_csv(f"./q_vs_q/results/run_{i}/shock_c/q_vs_q_schemeC.csv")

    # average prices
    avg_prices["shock_none"].append((df_shock_none.iat[0, 1], df_shock_none.iat[0, 3]))
    avg_prices["shock_a"].append((df_shock_a.iat[0, 1], df_shock_a.iat[0, 3]))
    avg_prices["shock_b"].append((df_shock_b.iat[0, 1], df_shock_b.iat[0, 3]))
    avg_prices["shock_c"].append((df_shock_c.iat[0, 1], df_shock_c.iat[0, 3]))

    # deltas
    deltas["shock_none"].append((df_shock_none.iat[0, 5], df_shock_none.iat[0, 6]))
    deltas["shock_a"].append((df_shock_a.iat[0, 5], df_shock_a.iat[0, 6]))
    deltas["shock_b"].append((df_shock_b.iat[0, 5], df_shock_b.iat[0, 6]))
    deltas["shock_c"].append((df_shock_c.iat[0, 5], df_shock_c.iat[0, 6]))

    # rpdis
    rpdis["shock_none"].append((df_shock_none.iat[0, 7], df_shock_none.iat[0, 8]))
    rpdis["shock_a"].append((df_shock_a.iat[0, 7], df_shock_a.iat[0, 8]))
    rpdis["shock_b"].append((df_shock_b.iat[0, 7], df_shock_b.iat[0, 8]))
    rpdis["shock_c"].append((df_shock_c.iat[0, 7], df_shock_c.iat[0, 8]))
```

## Shock None

### Average Prices

```python
series = avg_prices["shock_none"]

for x in series:
    print(f"{x[0]}\t{x[1]}")

print("-- Mean --")
print(f"{round(statistics.mean([x[0] for x in series]), 2)}\t{round(statistics.mean([x[1] for x in series]), 2)}")
print("-- St. Dev --")
print(f"{round(statistics.stdev([x[0] for x in series]), 2)}\t{round(statistics.stdev([x[1] for x in series]), 2)}")
```

### Deltas

```python
series = deltas['shock_none']

for x in series:
    print(f"{x[0]}\t{x[1]}")

print("-- Mean --")
print(f"{round(statistics.mean([x[0] for x in series]), 2)}\t{round(statistics.mean([x[1] for x in series]), 2)}")
print("-- St. Dev --")
print(f"{round(statistics.stdev([x[0] for x in series]), 2)}\t{round(statistics.stdev([x[1] for x in series]), 2)}")
```

```python
series = avg_prices["shock_c"]

for x in series:
    print(f"{x[0]}\t{x[1]}")

print("-- Mean --")
print(f"{round(statistics.mean([x[0] for x in series]), 2)}\t{round(statistics.mean([x[1] for x in series]), 2)}")
print("-- St. Dev --")
print(f"{round(statistics.stdev([x[0] for x in series]), 2)}\t{round(statistics.stdev([x[1] for x in series]), 2)}")
```
