# Imports

```python
import numpy as np
import pandas as pd
from scipy.stats import permutation_test
```

# Data Loading

```python
models = [
    "logit", # row 0
    "hotelling", # row 1
    "linear" # row 2
]

firm1 = "q"
firm2 = "dqn"

shock_none_data = {
    model: {
        "avg_prices": {
            firm1: [],
            firm2: []
        },
        "deltas": {
            firm1: [],
            firm2: []
        },
        "rpdis": {
            firm1: [],
            firm2: []
        }
    } for model in models
}

shock_a_data = {
    model: {
        "avg_prices": {
            firm1: [],
            firm2: []
        },
        "deltas": {
            firm1: [],
            firm2: []
        },
        "rpdis": {
            firm1: [],
            firm2: []
        }
    } for model in models
}

shock_b_data = {
    model: {
        "avg_prices": {
            firm1: [],
            firm2: []
        },
        "deltas": {
            firm1: [],
            firm2: []
        },
        "rpdis": {
            firm1: [],
            firm2: []
        }
    } for model in models
}

shock_c_data = {
    model: {
        "avg_prices": {
            firm1: [],
            firm2: []
        },
        "deltas": {
            firm1: [],
            firm2: []
        },
        "rpdis": {
            firm1: [],
            firm2: []
        }
    } for model in models
}

for i in range(1,8):
    shock_none_df = pd.read_csv(f"../results/run_{i}/shock_none/{firm1}_vs_{firm2}.csv")
    shock_a_df = pd.read_csv(f"../results/run_{i}/shock_a/{firm1}_vs_{firm2}_schemeA.csv")
    shock_b_df = pd.read_csv(f"../results/run_{i}/shock_b/{firm1}_vs_{firm2}_schemeB.csv")
    shock_c_df = pd.read_csv(f"../results/run_{i}/shock_c/{firm1}_vs_{firm2}_schemeC.csv")

    # Map each dataframe to its corresponding data storage dictionary
    shock_mappings = [
        (shock_none_df, shock_none_data),
        (shock_a_df, shock_a_data),
        (shock_b_df, shock_b_data),
        (shock_c_df, shock_c_data)
    ]
    
    for df, data_dict in shock_mappings:
        for idx, model in enumerate(models):
            # Access the row matching the current model by its index (0, 1, or 2)
            row = df.iloc[idx]
            
            # Dynamically look up columns based on firm name capitalization (e.g., "Q" and "PSO")
            f1_upper = firm1.upper()
            f2_upper = firm2.upper()
            
            # Append Average Prices
            data_dict[model]["avg_prices"][firm1].append(row[f"{f1_upper} Avg. Prices"])
            data_dict[model]["avg_prices"][firm2].append(row[f"{f2_upper} Avg. Prices"])
            
            # Append Extra-profits Δ (deltas)
            data_dict[model]["deltas"][firm1].append(row[f"{f1_upper} Δ"])
            data_dict[model]["deltas"][firm2].append(row[f"{f2_upper} Δ"])
            
            # Append RPDI
            data_dict[model]["rpdis"][firm1].append(row[f"{f1_upper} RPDI"])
            data_dict[model]["rpdis"][firm2].append(row[f"{f2_upper} RPDI"])
```

```python
def paired_mean_diff(x, y, axis=0):
    return np.mean(x - y, axis=axis)
```

# Logit

## Shock None

### Average Prices

```python
result = permutation_test(
    data=(shock_none_data['logit']['avg_prices'][firm1], shock_none_data['logit']['avg_prices'][firm2]), 
    statistic=paired_mean_diff, 
    permutation_type='samples',     # 'samples' ensures data remains paired
    alternative='two-sided',        # Options: 'two-sided', 'less', or 'greater'
    n_resamples=np.inf,             # Use np.inf for an exact test, or an integer like 9999 for larger data
    random_state=42                 # Sets a seed for reproducibility
)

print("--- Paired Permutation Test Results ---")
print(f"Observed Mean Difference ({firm1.upper()} - {firm2.upper()}): {result.statistic:.4f}")
print(f"Exact P-value: {result.pvalue:.4f}")

# Evaluate significance
alpha = 0.05
if result.pvalue <= alpha:
    print(f"Result is statistically significant (p <= {alpha}). Reject the null hypothesis. There is a significant difference between the two firms set prices.")
else:
    print(f"Result is not statistically significant (p > {alpha}). Fail to reject the null hypothesis. There is no significant difference between the two firms set prices.")
```

### Deltas

```python
result = permutation_test(
    data=(shock_none_data['logit']['deltas'][firm1], shock_none_data['logit']['deltas'][firm2]), 
    statistic=paired_mean_diff, 
    permutation_type='samples',     # 'samples' ensures data remains paired
    alternative='two-sided',        # Options: 'two-sided', 'less', or 'greater'
    n_resamples=np.inf,             # Use np.inf for an exact test, or an integer like 9999 for larger data
    random_state=42                 # Sets a seed for reproducibility
)

print("--- Paired Permutation Test Results ---")
print(f"Observed Mean Difference ({firm1.upper()} - {firm2.upper()}): {result.statistic:.4f}")
print(f"Exact P-value: {result.pvalue:.4f}")

# Evaluate significance
alpha = 0.05
if result.pvalue <= alpha:
    print(f"Result is statistically significant (p <= {alpha}). Reject the null hypothesis. There is a significant difference between the two firms deltas.")
else:
    print(f"Result is not statistically significant (p > {alpha}). Fail to reject the null hypothesis. There is no significant difference between the two firms deltas.")
```
