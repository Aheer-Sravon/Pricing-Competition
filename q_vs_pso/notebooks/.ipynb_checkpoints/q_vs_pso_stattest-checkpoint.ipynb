# Imports

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import permutation_test
```

```python
def paired_mean_statistic(x, y, axis=0):
    return np.mean(x - y, axis=axis)
```

# Logit

```python
market = "logit"
```

## No Shock

### load data

```python
q_prices, q_deltas, q_rpdis = [{
                                    "logit": [],
                                    "hotelling": [],
                                    "linear": []
                                } for i in range(3)]

pso_prices, pso_deltas, pso_rpdis = [{
                                    "logit": [],
                                    "hotelling": [],
                                    "linear": []
                                } for i in range(3)]

for i in range(1, 8):
    df = pd.read_csv(f"../results/run_{i}/shock_none/q_vs_pso.csv")

    # Q
    q_prices[market].append(df.iloc[0, 1])
    q_deltas[market].append(df.iloc[0, 5])
    q_rpdis[market].append(df.iloc[0, 7])

    # PSO
    pso_prices[market].append(df.iloc[0, 3])
    pso_deltas[market].append(df.iloc[0, 6])
    pso_rpdis[market].append(df.iloc[0, 8])
```

### test

```python
x = np.array(q_prices[market])
y = np.array(pso_prices[market])

res = permutation_test(
        data=(x, y), 
        statistic=paired_mean_statistic, 
        permutation_type='samples', 
        alternative='two-sided', 
        n_resamples=np.inf
    )
is_significant = "Yes" if res.pvalue <= 0.05 else "No"

print(f"{market.capitalize():<18} | {res.statistic:<30.4f} | {res.pvalue:<15.4f} | {is_significant}")
```
