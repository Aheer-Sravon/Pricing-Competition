# Imports

```python
import numpy as np
import pandas as pd
from scipy.stats import kruskal
import scikit_posthocs as sp
```

# Logit, Firm 1

## Data Loading

```python
model = 'logit' # row number = 0

avg_prices = {
    "shock_none": [],
    "shock_a": [],
    "shock_b": [],
    "shock_c": []
}

deltas = {
    "shock_none": [],
    "shock_a": [],
    "shock_b": [],
    "shock_c": []
}

rpdis = {
    "shock_none": [],
    "shock_a": [],
    "shock_b": [],
    "shock_c": []
}

for i in range(1, 8):
    # load the dataframes
    df_shock_none = pd.read_csv(f"../results/run_{i}/shock_none/q_vs_q.csv")
    df_shock_a = pd.read_csv(f"../results/run_{i}/shock_a/q_vs_q_schemeA.csv")
    df_shock_b = pd.read_csv(f"../results/run_{i}/shock_b/q_vs_q_schemeB.csv")
    df_shock_c = pd.read_csv(f"../results/run_{i}/shock_c/q_vs_q_schemeC.csv")

    # average prices
    avg_prices["shock_none"].append(df_shock_none.iat[0, 1])
    avg_prices["shock_a"].append(df_shock_a.iat[0, 1])
    avg_prices["shock_b"].append(df_shock_b.iat[0, 1])
    avg_prices["shock_c"].append(df_shock_c.iat[0, 1])

    # deltas
    deltas["shock_none"].append(df_shock_none.iat[0, 5])
    deltas["shock_a"].append(df_shock_a.iat[0, 5])
    deltas["shock_b"].append(df_shock_b.iat[0, 5])
    deltas["shock_c"].append(df_shock_c.iat[0, 5])

    # rpdis
    rpdis["shock_none"].append(df_shock_none.iat[0, 7])
    rpdis["shock_a"].append(df_shock_a.iat[0, 7])
    rpdis["shock_b"].append(df_shock_b.iat[0, 7])
    rpdis["shock_c"].append(df_shock_c.iat[0, 7])
```

## Kruskal-Wallis Test - Average Prices

```python
h_stat, p_val = kruskal(
    avg_prices["shock_none"], 
    avg_prices["shock_a"], 
    avg_prices["shock_b"], 
    avg_prices["shock_c"]
)
n = 4 * 7 # total number of observations across all groups
epsilon_squared = h_stat / (n - 1)

print("Kruskal-Wallis Test on Average Prices")
print(f"H-statistic: {h_stat:.2f}, P-value: {p_val:.8f}, epsilon_squared: {epsilon_squared:.4f}\n")
```

## Post-Hoc Testing - Average Prices

```python
dunn_matrix = sp.posthoc_dunn(
    [
        avg_prices["shock_none"], 
        avg_prices["shock_a"], 
        avg_prices["shock_b"], 
        avg_prices["shock_c"]
    ], p_adjust='holm'
)

print(dunn_matrix)
```

## Kruskal-Wallis Test - Deltas

```python
h_stat, p_val = kruskal(
    deltas["shock_none"], 
    deltas["shock_a"], 
    deltas["shock_b"], 
    deltas["shock_c"]
)
n = 4 * 7 # total number of observations across all groups
epsilon_squared = h_stat / (n - 1)

print("Kruskal-Wallis Test on Deltas")
print(f"H-statistic: {h_stat:.2f}, P-value: {p_val:.8f}, epsilon_squared: {epsilon_squared:.4f}\n")
```

## Post-Hoc Testing - Deltas

```python
dunn_matrix = sp.posthoc_dunn(
    [
        deltas["shock_none"], 
        deltas["shock_a"], 
        deltas["shock_b"], 
        deltas["shock_c"]
    ], p_adjust='holm'
)

print(dunn_matrix)
```

## Kruskal-Wallis Test - RPDIs

```python
h_stat, p_val = kruskal(
    rpdis["shock_none"], 
    rpdis["shock_a"], 
    rpdis["shock_b"], 
    rpdis["shock_c"]
)
n = 4 * 7 # total number of observations across all groups
epsilon_squared = h_stat / (n - 1)

print("Kruskal-Wallis Test on RPDIs")
print(f"H-statistic: {h_stat:.2f}, P-value: {p_val:.8f}, epsilon_squared: {epsilon_squared:.4f}\n")
```

## Post-Hoc Testing - RPDIs

```python
dunn_matrix = sp.posthoc_dunn(
    [
        rpdis["shock_none"], 
        rpdis["shock_a"], 
        rpdis["shock_b"], 
        rpdis["shock_c"]
    ], p_adjust='holm'
)

print(dunn_matrix)
```
